```python
from ultralytics import YOLOv10

MODEL_PATH = 'yolov10n.pt'
model = YOLOv10(MODEL_PATH)
```

```python
model.info()
```

```python
YAML_PATH = '/content/yolov10/safety_helmet_dataset/data.yaml'
EPOCHS = 30
IMG_SIZE = 320
BATCH_SIZE = 32

model.train(data=YAML_PATH,
            epochs=EPOCHS,
            imgsz=IMG_SIZE,
            batch=BATCH_SIZE)
```

```python
TRAINED_MODEL_PATH = '/content/yolov10/runs/detect/train/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)

results = model.predict(data=YAML_PATH,
                        imgsz=IMG_SIZE,
                        split='test')
```

```python
TRAINED_MODEL_PATH = '/content/yolov10/runs/detect/train/weights/last.pt'
model = YOLOv10(TRAINED_MODEL_PATH)

results = model.predict(data=YAML_PATH,
                        imgsz=IMG_SIZE,
                        split='test')
```

```python
from google.colab.patches import cv2_imshow

IMAGE_URL = 'https://nonbaohiemthangloi.com.vn/uploads/news/11_2018/non-bao-hiem-thang-loi-021.jpeg'
CONF_THRESHOLD = 0.5
results = model.predict(source=IMAGE_URL,
                        imgsz=IMG_SIZE,
                        conf=CONF_THRESHOLD)
annotated_img = results[0].plot()

cv2_imshow(annotated_img)
```
